```python
from sentence_transformers import SentenceTransformer

# Load a model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Encode sentences
sentences = ["This is an example sentence.", "Each sentence is converted into a vector."]
embeddings = model.encode(sentences)

print(embeddings)
```

```python
import pandas as pd
from transformers import pipeline

# Load your CSV file into a DataFrame
 # Replace with your actual file path
df = pd.read_csv('data/adae.csv')

# Convert all DataFrame columns to string type
df = df.astype(str)

# Initialize the TAPAS model for question answering
pipe = pipeline("table-question-answering", model="microsoft/tapex-large-finetuned-wtq")

# Example question
question = "list of patients with on going status"  # Modify this based on your data

# Convert the DataFrame to a list of dictionaries
table = df.to_dict(orient='records')  # Each record (row) becomes a dictionary
columns = df.columns.tolist()  # Get the column names

# Call the TAPAS model
result = pipe(table=table, query=question)
print(result)
# Access the answer based on the result structure
if result:
    # Check if there's an answer in the result
    answer = result[0].get('answer') if isinstance(result, list) and len(result) > 0 else None
    if answer:
        print("Answer:", answer)
    else:
        print("No answer found.")
else:
    print("No result returned.")
```
